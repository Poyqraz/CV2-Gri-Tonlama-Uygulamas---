```
import cv2
from matplotlib import pyplot as plt
```

```
# Define our imshow function - İmshow işlevini tanımlayın
def imshow(title = "Image", image = None, size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()

# Download and unzip our images - Resimleri indirin ve açın
# !wget https://moderncomputervision.s3.eu-west-2.amazonaws.com/images.zip
# !unzip -qq images.zip
```

```
# Load our input image - Kullanmak istenen resmi yükle.
image = cv2.imread('/content/drive/MyDrive/test_resimleri/sturmtiger.jpg')

imshow("sturmtiger", image)   # Drive'de bulunan örnek resmin kayıtlı olduğu isim.
```

```
image.shape[:2] # (1200, 1920) (height, width)
def imshow(title = "", image = None, size = 10):

    # The line below is changed from w, h to h, w  -- Burada w,h ' h,w ye çevriliyor.
    h, w = image.shape[0], image.shape[1]
    aspect_ratio = w/h

    plt.figure(figsize=(size * aspect_ratio, size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()

imshow("sturmtiger", image)
```

```
# We use cvtColor to convert to grayscale  -- Gri tonlamaya çevirmek için opencv cvtColor metodu kullanılıyor. (150 renk çeşidi var)
# It takes 2 arguments, the first being the input image -- İlki giriş görüntüsü olmak üzere 2 argümanımız var.
# The second being the color space conversion code  --  İkincisi ise renk uzayını dönüştürme kodudur.

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

imshow("Converted to Grayscale", gray_image)
```

### **Grayscale Image Dimensions**

Remember RGB color images have 3 dimensions, one for each primary color. Grayscale just has 1, which is the intensity of gray.
--
RGB renkli görüntülerin her ana renk için bir tane olmak üzere 3 boyutu olduğunu unutmayın. Gri tonlama sadece 1 tanesine sahiptir. Bu ise grinin yoğunluğudur.
